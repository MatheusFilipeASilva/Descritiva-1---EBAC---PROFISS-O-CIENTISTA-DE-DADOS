```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline


sinasc_raw = pd.read_csv('sinasc_RO_2019.csv')
sinasc_raw.shape
colunas_alvo = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE', 'CONSULTAS']
sinasc_toplot = sinasc_raw[colunas_alvo]
sinasc_toplot.shape
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.rc('figure',figsize=(8,6))
fig, axes = plt.subplots(1, 1)
sns.barplot(ax = axes, x = 'CONSULTAS', y = 'APGAR5', data = sinasc_toplot)

# Este gra´fico nos mostra um ponto que merece certo destaque: a maior quantidade de consultas não parece resultar em um aumento 100% proporcional no
# APGAR-5, que é um indicador da saúde do bebê, embora estejamos moderadamente dentro da margem de erro. A ligeira discrepância negativa de 4 para 9 
# consultas, no entanto, obviamente, não se deve ao fato de que mais consultas geram um bebê menos saudável, mas sim do fato de que, muito provavelmente
# mães que desde o começo da gestação já sabem que sua gestação é de risco, realizam muito mais consultas do que as que estão na expectativa de um bebê
# saudável.
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(10,10))
fig, axes = plt.subplots(2,2)

#Para esta tarefa, usaremos o countplot:
print(sinasc_toplot['GESTACAO'].value_counts())
sns.countplot(ax = axes[0, 0], x = 'SEXO', data = sinasc_toplot)
sns.countplot(ax = axes[0, 1], x = 'GESTACAO', data = sinasc_toplot)
# Neste ponto, foi necessário fazer uma alteração nos tick labels, pois os originais eram muito grandes, e ficariam sobrepostos em caso sem rotação, e
# com rotação, ficariam parcialmente cortados...
axes[0, 1].set_xticklabels(['37 a 41', '42+', '32 a 36', '22 a 27', '28 a 31', '22 -'], rotation = 90)
sns.countplot(ax = axes[1, 0], x = 'ESTCIVMAE', data = sinasc_toplot)
sns.countplot(ax = axes[1, 1], x = 'CONSULTAS', data = sinasc_toplot)

```
